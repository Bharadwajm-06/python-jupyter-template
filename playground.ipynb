```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = x ** 2  # Square of x

# Create the plot
plt.plot(x, y, label="y = x²", color="b")

# Customize the plot
plt.title("Line Plot of y = x²")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)  # Add grid
plt.legend()  # Show legend

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'purple']  # Different colors for each bar

# Create the bar plot
plt.bar(categories, values, color=colors)

# Customize the plot
plt.title("Bar Plot of Categories and Values")
plt.xlabel("Categories")
plt.ylabel("Values")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create scatter plot
plt.scatter(x, y, color='blue', marker='o')

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='right')

# Customize the plot
plt.title("Scatter Plot with Annotations")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)  # Add grid for better readability

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['blue', 'red', 'green', 'orange']  # Different colors for each section

# Create pie chart
plt.figure(figsize=(6,6))  # Set figure size
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)

# Add title
plt.title("Programming Language Popularity")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for each plot
x = np.linspace(0, 10, 100)
y = x ** 2

bar_categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]
pie_colors = ['blue', 'red', 'green', 'orange']

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot - Line Chart
axs[0, 0].plot(x, y, label="y = x²", color="b")
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X values")
axs[0, 0].set_ylabel("Y values")
axs[0, 0].grid(True)
axs[0, 0].legend()

# Second subplot - Bar Chart
axs[0, 1].bar(bar_categories, bar_values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Third subplot - Scatter Plot
axs[1, 0].scatter(scatter_x, scatter_y, color='blue', marker='o')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X values")
axs[1, 0].set_ylabel("Y values")
axs[1, 0].grid(True)

# Fourth subplot - Pie Chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=pie_colors, startangle=140)
axs[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()
plt.show()

```
