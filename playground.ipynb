```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = x ** 2  # Square of x

# Create the plot
plt.plot(x, y, label="y = x²", color="b")

# Customize the plot
plt.title("Line Plot of y = x²")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)  # Add grid
plt.legend()  # Show legend

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create scatter plot
plt.scatter(x, y, color='blue', marker='o')

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='right')

# Customize the plot
plt.title("Scatter Plot with Annotations")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.grid(True)  # Add grid for better readability

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt

# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]
colors = ['blue', 'red', 'green', 'orange']  # Different colors for each section

# Create pie chart
plt.figure(figsize=(6,6))  # Set figure size
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)

# Add title
plt.title("Programming Language Popularity")

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for each plot
x = np.linspace(0, 10, 100)
y = x ** 2

bar_categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]
pie_colors = ['blue', 'red', 'green', 'orange']

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot - Line Chart
axs[0, 0].plot(x, y, label="y = x²", color="b")
axs[0, 0].set_title("Line Chart")
axs[0, 0].set_xlabel("X values")
axs[0, 0].set_ylabel("Y values")
axs[0, 0].grid(True)
axs[0, 0].legend()

# Second subplot - Bar Chart
axs[0, 1].bar(bar_categories, bar_values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")
axs[0, 1].set_xlabel("Categories")
axs[0, 1].set_ylabel("Values")

# Third subplot - Scatter Plot
axs[1, 0].scatter(scatter_x, scatter_y, color='blue', marker='o')
axs[1, 0].set_title("Scatter Plot")
axs[1, 0].set_xlabel("X values")
axs[1, 0].set_ylabel("Y values")
axs[1, 0].grid(True)

# Fourth subplot - Pie Chart
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=pie_colors, startangle=140)
axs[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

def check_imports():
    try:
        import numpy
        import matplotlib
    except ImportError as e:
        print(f"Error: Missing library - {e}")
        return False
    return True

def validate_data(data):
    if not isinstance(data, (list, np.ndarray)):
        raise ValueError("Data should be a list or numpy array")
    if len(data) == 0:
        raise ValueError("Data should not be empty")
    return True

def line_plot():
    try:
        x = np.linspace(0, 10, 100)
        y = x ** 2
        validate_data(x)
        validate_data(y)
        plt.plot(x, y, label="y = x^2")
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")
        plt.title("Line Plot")
        plt.legend()
        plt.grid()
        plt.show()
    except Exception as e:
        print(f"Error in line plot: {e}")

def bar_plot():
    try:
        categories = ["A", "B", "C", "D"]
        values = [10, 20, 15, 25]
        validate_data(values)
        plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])
        plt.xlabel("Categories")
        plt.ylabel("Values")
        plt.title("Bar Plot")
        plt.show()
    except Exception as e:
        print(f"Error in bar plot: {e}")

def scatter_plot():
    try:
        x = np.random.rand(10)
        y = np.random.rand(10)
        validate_data(x)
        validate_data(y)
        plt.scatter(x, y, color='purple')
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")
        plt.title("Scatter Plot")
        plt.grid()
        plt.show()
    except Exception as e:
        print(f"Error in scatter plot: {e}")

def pie_chart():
    try:
        labels = ['A', 'B', 'C', 'D']
        sizes = [25, 35, 20, 20]
        validate_data(sizes)
        plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightcoral', 'lightgreen'])
        plt.title("Pie Chart")
        plt.show()
    except Exception as e:
        print(f"Error in pie chart: {e}")

def histogram():
    try:
        data = np.random.randn(1000)
        validate_data(data)
        plt.hist(data, bins=20, color='blue', edgecolor='black')
        plt.xlabel("Value")
        plt.ylabel("Frequency")
        plt.title("Histogram")
        plt.show()
    except Exception as e:
        print(f"Error in histogram: {e}")

def subplot_example():
    try:
        fig, axes = plt.subplots(2, 2, figsize=(10, 8))
        x = np.linspace(0, 10, 100)
        y = x ** 2
        validate_data(x)
        validate_data(y)
        
        # Line plot
        axes[0, 0].plot(x, y)
        axes[0, 0].set_title("Line Plot")
        
        # Bar plot
        categories = ["A", "B", "C", "D"]
        values = [10, 20, 15, 25]
        axes[0, 1].bar(categories, values, color='red')
        axes[0, 1].set_title("Bar Plot")
        
        # Scatter plot
        scatter_x = np.random.rand(10)
        scatter_y = np.random.rand(10)
        validate_data(scatter_x)
        validate_data(scatter_y)
        axes[1, 0].scatter(scatter_x, scatter_y, color='purple')
        axes[1, 0].set_title("Scatter Plot")
        
        # Pie chart
        sizes = [25, 35, 20, 20]
        validate_data(sizes)
        axes[1, 1].pie(sizes, labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightcoral', 'lightgreen'])
        axes[1, 1].set_title("Pie Chart")
        
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Error in subplot: {e}")

if __name__ == "__main__":
    if check_imports():
        line_plot()
        bar_plot()
        scatter_plot()
        pie_chart()
        histogram()
        subplot_example()

```
