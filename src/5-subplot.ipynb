### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def validate_data(data):
    if not isinstance(data, list) and not isinstance(data, np.ndarray):
        raise ValueError("Data should be a list or numpy array")
    if len(data) == 0:
        raise ValueError("Data should not be empty")
    return True

def subplot_example():
    try:
        fig, axes = plt.subplots(2, 2, figsize=(10, 8))
        x = np.linspace(0, 10, 100)
        y = x ** 2
        validate_data(x)
        validate_data(y)
        
        # Line plot
        axes[0, 0].plot(x, y)
        axes[0, 0].set_title("Line Plot")
        
        # Bar plot
        categories = ["A", "B", "C", "D"]
        values = [10, 20, 15, 25]
        axes[0, 1].bar(categories, values, color='red')
        axes[0, 1].set_title("Bar Plot")
        
        # Scatter plot
        scatter_x = np.random.rand(10)
        scatter_y = np.random.rand(10)
        validate_data(scatter_x)
        validate_data(scatter_y)
        axes[1, 0].scatter(scatter_x, scatter_y, color='purple')
        axes[1, 0].set_title("Scatter Plot")
        
        # Pie chart
        sizes = [25, 35, 20, 20]
        validate_data(sizes)
        axes[1, 1].pie(sizes, labels=['A', 'B', 'C', 'D'], autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightcoral', 'lightgreen'])
        axes[1, 1].set_title("Pie Chart")
        
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Error in subplot: {e}")

subplot_example()
```
