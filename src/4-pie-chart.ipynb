### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def validate_data(data):
    if not isinstance(data, list):
        raise ValueError("Data should be a list")
    if len(data) == 0:
        raise ValueError("Data should not be empty")
    return True

def pie_chart():
    try:
        labels = ['Python', 'Java', 'C++', 'JavaScript']
        sizes = [40, 25, 20, 15]
        validate_data(sizes)

        plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['gold', 'lightblue', 'lightcoral', 'lightgreen'])
        plt.title("Pie Chart")
        plt.show()
    except Exception as e:
        print(f"Error in pie chart: {e}")

pie_chart()
```
