### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def validate_data(data):
    if not isinstance(data, list):
        raise ValueError("Data should be a list")
    if len(data) == 0:
        raise ValueError("Data should not be empty")
    return True

def bar_plot():
    try:
        categories = ["A", "B", "C", "D"]
        values = [10, 20, 15, 25]
        validate_data(values)
        
        plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])
        plt.xlabel("Categories")
        plt.ylabel("Values")
        plt.title("Bar Plot")
        plt.show()
    except Exception as e:
        print(f"Error in bar plot: {e}")

bar_plot()
```
