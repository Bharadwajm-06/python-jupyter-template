### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def validate_data(data):
    if not isinstance(data, (list, np.ndarray)):
        raise ValueError("Data should be a list or numpy array")
    if len(data) == 0:
        raise ValueError("Data should not be empty")
    return True

def line_plot():
    try:
        x = np.linspace(0, 10, 100)
        y = x ** 2
        validate_data(x)
        validate_data(y)
        
        plt.plot(x, y, label="y = x^2")
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")
        plt.title("Line Plot")
        plt.legend()
        plt.grid()
        plt.show()
    except Exception as e:
        print(f"Error in line plot: {e}")

line_plot()
```
