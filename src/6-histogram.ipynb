### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def validate_data(data):
    if not isinstance(data, np.ndarray):
        raise ValueError("Data should be a numpy array")
    if len(data) == 0:
        raise ValueError("Data should not be empty")
    return True

def histogram():
    try:
        data = np.array([1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5])
        validate_data(data)

        plt.hist(data, bins=5, color='blue', edgecolor='black')
        plt.xlabel("Value")
        plt.ylabel("Frequency")
        plt.title("Histogram")
        plt.show()
    except Exception as e:
        print(f"Error in histogram: {e}")

histogram()
```
