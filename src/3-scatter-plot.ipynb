### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import numpy as np
import matplotlib.pyplot as plt

def validate_data(data):
    if not isinstance(data, np.ndarray):
        raise ValueError("Data should be a numpy array")
    if len(data) == 0:
        raise ValueError("Data should not be empty")
    return True

def scatter_plot():
    try:
        x = np.array([1, 2, 3, 4, 5])
        y = np.array([2, 4, 6, 8, 10])
        validate_data(x)
        validate_data(y)
        
        plt.scatter(x, y, color='purple')
        for i, txt in enumerate(y):
            plt.annotate(f"({x[i]}, {y[i]})", (x[i], y[i]), textcoords="offset points", xytext=(0, 5), ha='center')
        
        plt.xlabel("X-axis")
        plt.ylabel("Y-axis")
        plt.title("Scatter Plot with Annotations")
        plt.grid()
        plt.show()
    except Exception as e:
        print(f"Error in scatter plot: {e}")

scatter_plot()
```
